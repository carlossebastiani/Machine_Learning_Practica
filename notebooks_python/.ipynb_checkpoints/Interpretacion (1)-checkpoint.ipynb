```python
filename = '../modelos/randomforest.sav'
clf=joblib.load(filename)
```

Nos quedamos con los primeros 15 variables explicativas que nos da la explicabilidad global del arbol de desición

```python
variable_importance=pd.DataFrame({'Variable':X_train.columns,'Importance':clf.feature_importances_}).sort_values('Importance', ascending=False).head(15)
variable_importance['Cumsum']=np.cumsum(variable_importance.Importance)
```

```python
plt.figure(figsize=(12, 6), dpi=80)
plt.bar(variable_importance.Variable, variable_importance.Importance,color='peru')
plt.step(variable_importance.Variable, variable_importance.Cumsum, where='mid',label='Cumulative explained variance',color='black')

plt.show()
```
