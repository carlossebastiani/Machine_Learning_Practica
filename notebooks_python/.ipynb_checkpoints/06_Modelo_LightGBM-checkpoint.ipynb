## Métricas

**Matriz de confusiones** solamente ha acertado un 19.2 % de los 1s. <br />
**Curva de ROC** vemos que es mejor que varios modelos, pues el área bajo la curva no es tan baja como en otros. Sabemos que mientras más se aleje la curva de ROC de la diagonal, quiere decir que predice mejor los verdaderos positivos y que disminuye el error de estos. Utilizando las métricas, podemos sacar el treshold que nos ayudará a cambiar el punto en el que vamos a decidir si efectivamente tomamos la probabilidad como un 1 o como un 0. Esto hace que aumentemos el ratio de verdaderos positivos considerablemente. <br />
**La curva de ganancia** , nos indica que porcentaje de la muestra tenemos que tomar para poder tomar a los 1s que logramos predecir. Vemos que en este caso tomando un 20% de la muestra, llegamos a predecir el 60% de los positivos. Viendo los modelos anteriores, esto está en línea, a excepción del arbol de desición que es más bajo. <br />
**La curva lift**, Nuetra curva de lift no es demasiado buena en este modelo, comparada por ejemplo contra la del random forest. Pues esta nos da que tanto impacto tendríamos en la deteccion de unos, untiliando nuestro modelo versus un modelo random, según se aumenta la muestra. En este caso, vemos que utilizando el modelo, si usamos el 10% de la población, somos capaces de detectar 10 unos más que si no lo usasemos. 
 
***Accuracy** elevado pero un **recall** muy bajo, lo que es indicador que del total de la muestra no detecta muchos accidentes mortales.

Sin embargo, cuando actualizamos el treshold y el F1 score, el recall que son las metricas que nos interesan, mejoran considerablemente.
