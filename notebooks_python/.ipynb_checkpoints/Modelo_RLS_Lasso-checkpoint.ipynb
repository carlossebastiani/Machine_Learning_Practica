### LASSO

En vista que, como se puede comrpobar abajo, nuestro modelo Lasso se queda con el total de las variables, hemos decidido no eliminar ninguna en nuestros modelos.

```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn import metrics
from sklearn.metrics import accuracy_score, plot_confusion_matrix, confusion_matrix, roc_curve, recall_score,precision_score
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectFromModel
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
#Creamos la función para sacar métricas
def print_metrics(y_test,y_pred):
    print("------------------------------------------")
    print("The accuracy score is: ", round(accuracy_score(y_test, y_pred),3))
    print("------------------------------------------")
    fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred)
    print("The F1 score is: ",round(metrics.auc(fpr, tpr),3))
    print("------------------------------------------")
    print("The recall is: ",round(recall_score(y_test,y_pred),3))
    print("------------------------------------------")
    print("The precision score is: ",round(precision_score(y_test,y_pred),3))
```

```python
# Cargamos los data sets y los ponemos en el formato correcto
X_train = pd.read_parquet('../PreprocessingSampled2/X_trainSampled2')
X_test = pd.read_parquet('../PreprocessingSampled2/X_test')
y_train=pd.read_parquet('../PreprocessingSampled2/y_trainSampled2')
y_test=pd.read_csv('../PreprocessingSampled2/y_test').set_index('Unnamed: 0')
```

```python
# Cargamos el modelo
filename = '../modelos/lasso.sav'
reg = joblib.load(filename)
```

## __No correr la siguiente celda si el modelo ha sido cargado__

```python
#Entrenamos el modelo con parámetros por defecto
reg = LogisticRegression()
reg.fit(X_train,y_train)
```

```python
# Guardamos nuestro modelo
# filename = '../modelos/lasso.sav'
# joblib.dump(reg, filename)
```

```python
# Predicciones
y_pred=reg.predict(X_test)
y_pred_proba=reg.predict_proba(X_test)
```

```python
print_metrics(y_test,y_pred)
```

```python
# Creamos las matrices de correlación
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(7,7))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Realidad');
cm2=[[cm[0][0]/sum(cm[0]),cm[0][1]/sum(cm[0])],[cm[1][0]/sum(cm[1]),cm[1][1]/sum(cm[1])]]
plt.figure(figsize=(7,7))
sns.heatmap(cm2, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Realidad')
```

```python
# Creamos la curva ROC
yhat = y_pred_proba[:, 1]

# Curva roc
fpr, tpr, thresholds = roc_curve(y_test, yhat)

# gmeans
gmeans = np.sqrt(tpr * (1-fpr))
ix = np.argmax(gmeans)
print('Best Threshold=%f, G-Mean=%.3f' % (thresholds[ix], gmeans[ix]))

# Ploteamos la curva
plt.plot([0,1], [0,1], linestyle='--', label='No Skill')
plt.plot(fpr, tpr, marker='.', label='Logistic Regression')
plt.scatter(fpr[ix], tpr[ix], s=100, marker='o', color='black', label='Best')

# axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

```python
# Seleccionamos variables lasso
sel_lasso = SelectFromModel(LogisticRegression(C=1, penalty='l1', 
                                          solver='liblinear'), threshold = 0.08) 
sel_lasso.fit(X_train, y_train)
```

```python
sel_lasso.get_support()
selected_feat_lasso = X_train.columns[sel_lasso.get_support()]
selected_feat_lasso
```

```python
print('total features: {}'.format((X_train.shape[1])))
print('selected features: {}'.format(len(selected_feat_lasso)))
```

```python
# Coeficientes del modelo
# ==============================================================================
df_coeficientes_lasso = pd.DataFrame(
                        {'predictor': X_train.columns,
                         'coef': sel_lasso.estimator_.coef_.flatten()}
                  )

fig, ax = plt.subplots(figsize=(16, 3.84))
ax.stem(df_coeficientes_lasso.predictor, df_coeficientes_lasso.coef, markerfmt=' ')
plt.xticks(rotation=90, ha='right', size=10)
ax.set_xlabel('variable')
ax.set_ylabel('coeficientes')
ax.set_title('Coeficientes del modelo lasso');
```
