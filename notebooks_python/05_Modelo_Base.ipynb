```python
# Cargamos los data sets y los ponemos en el formato correcto
X_train = pd.read_parquet('../PreprocessingSampled2/X_trainSampled2')
X_test = pd.read_parquet('../PreprocessingSampled2/X_test')
y_train= pd.read_parquet('../PreprocessingSampled2/y_trainSampled2')
y_test= pd.read_csv('../PreprocessingSampled2/y_test').set_index('Unnamed: 0')
```

```python
# Cargamos nuestro modelo
filename = '../modelos/decision.sav'
clf = joblib.load(filename)
```

## __No correr la siguiente celda si el modelo se ha cargado correctamente__

```python
# Entrenamos nuestro modelo
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
y_pred_proba= clf.predict_proba(X_test)
```

## Métricas

Este ha sido el modelo que ha arrojado peores métricas.
- __Matriz de confusiones__ solamente ha acertado un 30 % de los 1s.
- __Curva de ROC__ vemos que casi traza una diagonal, lo que es lo peor que puede hacer, puesto que el True Positive Rate y el False Positive rate crecen casi en la misma porporción, lo que demuestra la aleatoriedad de este modelo.
- Mirando la __curva pr__ vemos que la precisión, es decir, de los que decimos que son 1s son realmente 1s, va cayendo muy rápidamente a medida que queremos aumentar en recall, es decir, coger más 1s del total de la muestra. No es un buen indicador.
- La __curva de ganancia__ , ocurre algo parecido. A medida que aumentamos el tamaño de la muestra, el número de 1s que acierta no crece. Es una línea casi diagonal, lo que nuevamente nos indica que no es un buen modelo.
- Por último, mirando la __curva lift__, vemos que el valor es bajísimo comparado por ejemplo, con Random Forest, que llegaba a 60. Esto quiere decir que el ratio entre 1s acertados y 1s del dataset es realmente bajo.
- __Accuracy__ elevado pero un __recall__ muy bajo, lo que es indicador que del total de la muestra no detecta muchos accidentes mortales.
